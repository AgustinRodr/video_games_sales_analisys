Creo un modelo de regresión lineal vacío utilizando la clase LinearRegression. Este modelo se utilizará posteriormente para ajustar los datos y realizar predicciones utilizando la relación lineal entre la variable independiente y la variable dependiente.



---
Utilizo el modelo de regresión lineal ajustado para realizar predicciones de la variable dependiente (y) basadas en los valores de la variable independiente (X). Las predicciones se almacenan en la variable y_pred para su posterior uso o análisis.

```python
# Realizar las predicciones
y_pred = model.predict(X)
```



---
Bbtiene los coeficientes de regresión y el intercepto del modelo, y calcula el R2 score como medidas de evaluación del rendimiento del modelo de regresión lineal.
```python
# Calcular el error cuadrado medio (MSE)
mse = mean_squared_error(y, y_pred)
# Obtener los coeficientes del modelo
coefficients = model.coef_
intercept = model.intercept_
# Calcular el coeficiente de determinación (R2 score)
r2 = r2_score(y, y_pred)
```

Este codigo realiza el análisis de Componentes Principales (PCA) en las columnas numéricas del DataFrame, reduciendo la dimensionalidad y proporcionando información sobre la varianza explicada por cada componente principal. La visualización final muestra la varianza explicada acumulada para evaluar la cantidad de información retenida al seleccionar diferentes números de componentes principales.

Los resultados del análisis de Componentes Principales (PCA) muestran lo siguiente:

Componente 1: El componente 1 explica aproximadamente el 79.26% de la varianza total en los datos.
Componente 2: El componente 2 explica aproximadamente el 18.87% de la varianza total en los datos.
En conjunto, estos dos componentes principales explican aproximadamente el 98.13% de la varianza total en los datos.

Esto significa que al reducir la dimensionalidad de los datos a solo dos componentes principales, se conserva una gran cantidad de información, ya que la varianza explicada acumulada es alta. Esto sugiere que los dos componentes principales capturan características importantes de los datos originales.

En conclusión, al aplicar PCA a las columnas numéricas del DataFrame, se logra una reducción de la dimensionalidad y se retiene una gran parte de la variabilidad en los datos al considerar los dos componentes principales. Esto puede ser útil para visualizar y analizar los datos en un espacio de menor dimensión.


---



```python
from sklearn.decomposition import PCA

# Seleccionar solo las columnas numéricas para PCA
numeric_columns = df1.select_dtypes(include=np.number)

# Crear una instancia de PCA con el número deseado de componentes principales
pca = PCA(n_components=2)

# Aplicar PCA a los datos numéricos
reduced_data = pca.fit_transform(numeric_columns)

# Crear un nuevo DataFrame con los componentes principales reducidos
df_reduced = pd.DataFrame(data=reduced_data, columns=["Component 1", "Component 2"])

# Mostrar los primeros registros del nuevo DataFrame
print(df_reduced.head())

# Varianza explicada por cada componente
explained_variance = pca.explained_variance_ratio_
print("Varianza explicada por cada componente:")
print(explained_variance)

# Varianza explicada acumulada
cumulative_variance = np.cumsum(explained_variance)
print("Varianza explicada acumulada:")
print(cumulative_variance)

# Gráfico de varianza explicada acumulada
plt.plot(range(1, len(cumulative_variance) + 1), cumulative_variance, marker='o')
plt.xlabel("Número de componentes principales")
plt.ylabel("Varianza explicada acumulada")
plt.title("Varianza explicada acumulada por componentes principales")
plt.show()
```



---

La tabla de comparación muestra una comparación entre los datos originales y los datos reconstruidos después de aplicar la técnica de Análisis de Componentes Principales (PCA). Cada columna representa una variable del conjunto de datos.

#####En resumen, la tabla de comparación muestra cómo los datos originales se transforman y reconstruyen después de aplicar PCA, lo que permite evaluar la efectividad de la técnica en la reducción de la dimensionalidad y la conservación de la información.
---



```python
# Visualización de los componentes principales
plt.scatter(df_reduced["Component 1"], df_reduced["Component 2"])
plt.xlabel("Componente 1")
plt.ylabel("Componente 2")
plt.title("Visualización de los componentes principales")
plt.show()

# Reconstrucción de los datos originales
reconstructed_data = pca.inverse_transform(reduced_data)
df_reconstructed = pd.DataFrame(data=reconstructed_data, columns=numeric_columns.columns)

# Comparación entre datos originales y datos reconstruidos
comparison_data = pd.concat([numeric_columns, df_reconstructed], axis=1)
print("Comparación entre datos originales y datos reconstruidos:")
print(comparison_data.head())
```

Error Cuadrático Medio (MSE): El valor del MSE es 2.91. MSE mide la magnitud promedio de los errores cuadráticos entre las predicciones y los valores reales. Cuanto menor sea el MSE, mejor será el rendimiento del modelo. En este caso, el MSE de 2.91 sugiere que el modelo tiene errores medios moderados en sus predicciones.

Raíz del Error Cuadrático Medio (RMSE): El valor del RMSE es 1.71. El RMSE es la raíz cuadrada del MSE y tiene las mismas unidades que la variable de destino. Un RMSE más bajo indica que las predicciones son más precisas. Un RMSE de 1.71 sugiere que las predicciones tienden a estar a una distancia de alrededor de 1.71 unidades de las observaciones reales.

Error Absoluto Medio (MAE): El valor del MAE es 0.52. El MAE representa la magnitud promedio de los errores absolutos entre las predicciones y los valores reales. Un MAE más bajo indica un mejor rendimiento del modelo. Con un MAE de 0.52, el modelo tiene errores medios bajos en sus predicciones, lo que es una señal positiva.

Coeficiente de Determinación (R^2): El valor del coeficiente de determinación R^2 es 0.31. R^2 mide la proporción de la varianza en la variable dependiente que es explicada por el modelo. Un valor de R^2 cercano a 1 indica un buen ajuste, mientras que un valor cercano a 0 indica que el modelo no explica bien la variabilidad de los datos. En este caso, un R^2 de 0.31 sugiere que el modelo explica solo el 31% de la variabilidad en los datos, lo que no es muy alto.

En resumen, el modelo de regresión parece tener un rendimiento moderado. El RMSE y el MAE son relativamente bajos, lo que indica que las predicciones tienden a ser cercanas a los valores reales en términos absolutos. Sin embargo, el valor de R2 sugiere que el modelo no explica una gran parte de la variabilidad en los datos, por lo que podría haber margen de mejora en el modelo o en la elección de características.

```python
from sklearn.linear_model import Ridge

# VALORES ALPHA A EVALUAR
alpha_values = [0.0001,0.001,0.01,0.1]
ridge_models = []

print('Model comparison on the TRAINING SET:')

for alpha in alpha_values:
  # INICIALIZAMOS MODELO RIDGE REGRESSION
  ridge_model = Ridge(alpha=alpha, fit_intercept=True)
  # ENTRENAMIENTO
  ridge_model.fit(x_train, y_train)
  ridge_models.append(ridge_model)
  # Y LO EVALUAMOS SOBRE LOS DATOS DE TRAINING UTILIZANDO MSE
  y_hat_train = ridge_model.predict(x_test)
  current_mse = mean_squared_error(y_hat_train, y_test)
  print('Ridge with alpha: {:.4f} \t MSE = {:.4f}'.format(alpha, current_mse))
```
